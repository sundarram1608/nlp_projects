```python
def load_data(data_csv, answers_csv, is_test=False):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass
    if is_test == True:
        data = pd.read_csv(data_csv).dropna()
        answers = pd.read_csv(answers_csv, header=None).rename(columns={0: "id", 1: "reason1", 2: "reason2", 3: "reason3"})
        final_return_data = pd.merge(data, answers, on="id")     
        return final_return_data   
    
    else:
        data = pd.read_csv(data_csv).dropna()
        answers = pd.read_csv(answers_csv, header=None).rename(columns={0: "id", 1: "reason1", 2: "reason2", 3: "reason3"})
        prefinal_return_data = pd.merge(data, answers, on="id")        
        
        final_return_data = prefinal_return_data.melt(
                                                        id_vars=["id", "FalseSent"],
                                                        value_vars=["reason1", "reason2", "reason3"],
                                                        value_name="reason"
                                                    )

        final_return_data.sort_values(by=["id","variable"], inplace=True)
        final_return_data = final_return_data.drop(columns=["variable"])

        return final_return_data
        
```

```python
def preprocess_data(examples, tokenizer, max_length, is_test=False):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass

    if is_test == True:
            
        tokenized_batch = tokenizer(
                                    examples['FalseSent'],
                                    padding=True,
                                    truncation=True,
                                    max_length=max_length
                                )
        return_value = {
                            "input_ids": tokenized_batch["input_ids"],
                            "attention_mask": tokenized_batch["attention_mask"]
                        }
        return return_value
    else: 
        tokenized_batch = tokenizer(
                                        examples['FalseSent'],
                                        padding=True,
                                        truncation=True,
                                        max_length=max_length
                                    )
        
        tokenized_reason = tokenizer(
                                        examples["reason"],
                                        padding=True,
                                        truncation=True,
                                        max_length=max_length
                                    )
        return_value = {
                            "input_ids": tokenized_batch["input_ids"],
                            "attention_mask": tokenized_batch["attention_mask"],
                            "labels": tokenized_reason["input_ids"]
                        }
        return return_value
```

```python
train_dataset = train_dataset.map(lambda x: preprocess_data(x, tokenizer, max_length), batched=True)
dev_dataset = dev_dataset.map(lambda x: preprocess_data(x, tokenizer, max_length), batched=True)
test_dataset = test_dataset.map(lambda x: preprocess_data(x, tokenizer, max_length, True), batched=True)
print("Train formated Dataset example:\n")
print(train_dataset[0])
print("\nTest formated Dataset example:\n")
print(test_dataset[0])
```

## Fine-tuning

In general, when using a `Trainer` to make predictions, it returns the logits for each class in the task. However, the `Seq2SeqTrainingArguments` class provides an option that allows the `Trainer` to generate sequences of tokens in the prediction. The `create_training_arguments` function must create the `Seq2SeqTrainingArguments` with that option and the hyperparamters passed as arguments. During the training, the model must be evaluated on the development set after every epoch. `TrainingArguments` should include this strategy.

> **Important!** By default, `Trainer` saves a checkpoint of the model every 500 training steps. For this assignment, avoid this behavior by setting `save_strategy="no"` when creating the `TrainingArguments`.

```python
def create_training_arguments(epochs, train_batch_size, learning_rate, output_dir):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass
    training_args = Seq2SeqTrainingArguments(
                                            output_dir=output_dir,
                                            eval_strategy="epoch",
                                            num_train_epochs=epochs,
                                            per_device_train_batch_size=train_batch_size,
                                            learning_rate=learning_rate,
                                            save_strategy= "no",
                                            predict_with_generate= True
                                            )
    return training_args
```

```python
train_args = create_training_arguments(epochs, train_batch_size, learning_rate, output_dir)
```

Next, you can create a `Trainer` object initializing the appropriate data collator.

```python
def create_trainer(model, train_args, train_dataset, dev_dataset, tokenizer):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass
    data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer)
    trainer = Seq2SeqTrainer(
                    model=model,
                    args=train_args,
                    train_dataset=train_dataset,
                    eval_dataset=dev_dataset,
                    tokenizer=tokenizer,
                    data_collator=data_collator
                )
    return trainer
```

```python
trainer = create_trainer(model, train_args, train_dataset, dev_dataset, tokenizer)
```

```python
trainer.train()
```

```python
def make_predictions(trainer, test_dataset, tokenizer):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass
    predictions = trainer.predict(test_dataset)
    final_predictions = tokenizer.batch_decode(predictions.predictions, skip_special_tokens=True)
    
    return final_predictions
```

The **Subtasks C** of **ComVE** is evaluated using the *bleu* metric. In this assignment, you will also evaluate using *rouge*. With `shrink_dataset` and `base_model` set to `True`, the expected scores are *0.216* and *0.446* for *bleu* and *rouge* respectively. With a full training run, i.e. with `shrink_dataset` and `base_model` set to `False`, the scores should be around *0.228* and *0.461*.

```python
def evaluate_prediction(test_data, metric):
    #
    #  REPLACE THE pass STATEMENT WITH YOUR CODE
    #
    # pass
    
    if "bleu" in metric.lower():
        bleu_evaluation = evaluate.load("bleu")
        results = bleu_evaluation.compute(predictions=test_data["prediction"].tolist(), references=test_data[["reason1", "reason2", "reason3"]].values.tolist())
        return results
    else:
        rouge_evaluation = evaluate.load("rouge")
        results = rouge_evaluation.compute(predictions=test_data["prediction"].tolist(), references=test_data["reason1"].tolist())
        return results
```

```python
evaluate_prediction(test_data, "bleu")
```

```python
evaluate_prediction(test_data, "rouge")
```
